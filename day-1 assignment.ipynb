#Question 1 plot a line plot between a and b

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
plt.plot(a,b)
plt.title('Line-Plot')
plt.xlabel('a value')
plt.ylabel('b value')
plt.show()
```

#Question 2 Plot a line plot showing the sales trend in company 1 and 2:

```python
days = [1,2,3,4,5,6,7]
sales_1 = [160,150,140,145,175,165,180]
sales_2 = [70,90,160,150,140,145,175]
```

```python
plt.plot(days, sales_1, color='blue', linewidth=3)
plt.plot(days, sales_2, color='red', linewidth=3)
plt.title('sales trends')
plt.xlabel('days')
plt.ylabel('sales')
plt.show()
```

#Question 3 Create a 3 by 3 subplots:
#multiple plots


```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
```

```python
plt.figure(figsize = (8, 8))
plt.subplot(3,3,1)
plt.plot(x,y1,color='red')
plt.title('subplot-1')

plt.subplot(3,3,2)
plt.plot(x,y2,color='blue')
plt.title('subplot-2')

plt.subplot(3,3,3)
plt.plot(x,y3,color='yellow')
plt.title('subplot-3')

plt.subplot(3,3,4)
plt.plot(x,y4,color='green')
plt.title('subplot-4')

plt.subplot(3,3,5)
plt.plot(x,y5,color='brown')
plt.title('subplot-5')

plt.show()
```
